```python
import pandas as pd
```

```python
df = pd.read_csv('../data/600000.csv',index_col=0)
```

```python
df.head()
```

```python
df['']
```

```python
df.sort_values(by='tradeDate',inplace=True)
```

```python
df.head()
```

```python
ma_list = [5, 20, 60]
```

```python
for  ma in ma_list:
    df['MA_' + str(ma)] = df['closePrice'].rolling(ma).mean()
```

```python
df.head()
```

```python
for  ma in ma_list: 
    df['EMA_' + str(ma)]= pd.DataFrame.ewm(df['closePrice'], halflife=ma).mean()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,16))
df.set_index('tradeDate',inplace=True)
df[['closePrice','MA_5','MA_20','MA_60']].plot()
plt.show()
```
