```python
import pandas as pd
import matplotlib.pyplot as plt
import pyecharts.options as opts
from pyecharts.charts import Line
import datetime
```

```python
today=datetime.datetime.now().strftime('%Y-%m-%d')
```

```python
df=pd.read_excel('close_price_median0611.xlsx')
```

```python
df['date']=df['date'].astype(str)
df.set_index('date',drop=True,inplace=True)
df['closePriceBond']=df['closePriceBond'].map(lambda x:round(x,1))
df['bondPremRatio']=df['bondPremRatio'].map(lambda x:round(x,1))
df['remainSize']=df['remainSize'].map(lambda x:round(x,1))
```

```python
# 无点图
index=df.index.values
c = (
    Line()
    .add_xaxis(list(index))
    .add_yaxis("价格中位数",
               df['closePriceBond'].tolist(),
               is_smooth=True,
               symbol_size=8,
#                markline_opts=opts.MarkLineOpts(
#                    data=[opts.MarkLineItem(type_="average", name="平均值")]),
#                        linestyle_opts=opts.LineStyleOpts(width=1,color='rgb(0, 0, 255)'),
               yaxis_index=0,
               label_opts=opts.LabelOpts(is_show=False),
               is_symbol_show=False,

              )
    .add_yaxis('溢价率中位数',
                         df['bondPremRatio'].tolist(),
               is_smooth=True,
               symbol_size=8,
#                markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="average", name="平均值"),
#                                                     opts.MarkLineItem(type_="min", name="最小值"),
#                                                     opts.MarkLineItem(type_="max", name="最大值")
#                                                     ]
#                                               ),
#                        linestyle_opts=opts.LineStyleOpts(width=1,color='rgb(255, 0, 0)'),
                          yaxis_index=1,
               label_opts=opts.LabelOpts(is_show=False),
               is_symbol_show=False,
                         ).extend_axis(
        yaxis=opts.AxisOpts(
            name="溢价率中位数",
            type_="value",
            min_=df['bondPremRatio'].min(),
            max_=df['bondPremRatio'].max(),
            position="right",
#             axisline_opts=opts.AxisLineOpts(
#                 linestyle_opts=opts.LineStyleOpts(color=colors[1])
#             ),
#             axislabel_opts=opts.LabelOpts(formatter="{value} ml"),
        )
    ).extend_axis(
        yaxis=opts.AxisOpts(
            name="价格中位数",
            type_="value",
#             min_=90,
#             max_=140,
            position="left",
#             axisline_opts=opts.AxisLineOpts(
#                 linestyle_opts=opts.LineStyleOpts(color=colors[1])
#             ),
#             axislabel_opts=opts.LabelOpts(formatter="{value} ml"),
        )
    ).set_colors(["blue","red"])
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
                                name='日期',
                                    splitline_opts=opts.SplitLineOpts(is_show=True)),
        yaxis_opts=opts.AxisOpts(
                                                min_=df['closePriceBond'].min(),
            max_=df['closePriceBond'].max(),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
#         splitline_opts=opts.SplitLineOpts(is_show=True)),
        title_opts=opts.TitleOpts(title=f"{today}中位数"),
    )
    .render("line_smooth0611-nopot.html")
)
```

```python

# 有点的数据
index=df.index.values
c = (
    Line()
    .add_xaxis(list(index))
    .add_yaxis("价格中位数",
               df['closePriceBond'].tolist(),
               is_smooth=True,
               symbol_size=8,
               markline_opts=opts.MarkLineOpts(
                   data=[opts.MarkLineItem(type_="average", name="平均值")]),
                       linestyle_opts=opts.LineStyleOpts(width=1,color='rgb(0, 0, 255)'),
               yaxis_index=0,
               label_opts=opts.LabelOpts(is_show=True),
               is_symbol_show=True,

              )
    .add_yaxis('溢价率中位数',
                         df['bondPremRatio'].tolist(),
               is_smooth=True,
               symbol_size=8,
               markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="average", name="平均值"),
                                                    opts.MarkLineItem(type_="min", name="最小值"),
                                                    opts.MarkLineItem(type_="max", name="最大值")
                                                    ]
                                              ),
#                        linestyle_opts=opts.LineStyleOpts(width=1,color='rgb(255, 0, 0)'),
                          yaxis_index=1,
               label_opts=opts.LabelOpts(is_show=True),
               is_symbol_show=True,
                         ).extend_axis(
        yaxis=opts.AxisOpts(
            name="溢价率中位数",
            type_="value",
            min_=df['bondPremRatio'].min(),
            max_=df['bondPremRatio'].max(),
            position="right",
#             axisline_opts=opts.AxisLineOpts(
#                 linestyle_opts=opts.LineStyleOpts(color=colors[1])
#             ),
#             axislabel_opts=opts.LabelOpts(formatter="{value} ml"),
        )
    ).extend_axis(
        yaxis=opts.AxisOpts(
            name="价格中位数",
            type_="value",
#             min_=90,
#             max_=140,
            position="left",
#             axisline_opts=opts.AxisLineOpts(
#                 linestyle_opts=opts.LineStyleOpts(color=colors[1])
#             ),
#             axislabel_opts=opts.LabelOpts(formatter="{value} ml"),
        )
    ).set_colors(["green","blue"]) # 统一设置颜色
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
                                name='日期',
                                    splitline_opts=opts.SplitLineOpts(is_show=True)),
        yaxis_opts=opts.AxisOpts(
                                                min_=df['closePriceBond'].min(),
            max_=df['closePriceBond'].max(),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
#         splitline_opts=opts.SplitLineOpts(is_show=True)),
        title_opts=opts.TitleOpts(title="{today}中位数"),
    )
    .render("line_smooth0611.html")
)
```
